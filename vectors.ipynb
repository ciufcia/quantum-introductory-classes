### üìê Wektor / Vector



#### üá¨üáß English:
A **vector** is a mathematical object characterized by its **magnitude** (length), **direction**, and **sense**.  
In Euclidean space, a vector is represented as a **directed line segment** connecting two points: an initial point and a terminal point.  
Vectors are fundamental in mathematics, physics, and engineering for describing quantities such as **displacement**, **velocity**, and **force**.

**Components of a vector:**
- **Magnitude (length):** The distance between the initial and terminal points.
- **Direction:** The straight line along which the vector lies.
- **Sense:** Indicates the direction along the line that the vector points.

**Example:**  
A vector **(3, 4)** in 2D space represents a displacement of 3 units to the right and 4 units upward.

---

#### üáµüá± Polski:
**Wektor** to obiekt matematyczny charakteryzujƒÖcy siƒô **modu≈Çem** (d≈Çugo≈õciƒÖ), **kierunkiem** oraz **zwrotem**.  
W przestrzeni euklidesowej, wektor jest reprezentowany jako **odcinek skierowany** ≈ÇƒÖczƒÖcy dwa punkty: poczƒÖtkowy i ko≈Ñcowy.  
Wektory sƒÖ kluczowe w matematyce, fizyce oraz in≈ºynierii do opisu wielko≈õci takich jak **przemieszczenie**, **prƒôdko≈õƒá** czy **si≈Ça**.

**Elementy wektora:**
- **Modu≈Ç (d≈Çugo≈õƒá):** Odleg≈Ço≈õƒá miƒôdzy punktem poczƒÖtkowym a ko≈Ñcowym.
- **Kierunek:** Linia prosta, wzd≈Çu≈º kt√≥rej le≈ºy wektor.
- **Zwrot:** Okre≈õla, w kt√≥rƒÖ stronƒô wzd≈Çu≈º kierunku wskazuje wektor.

**Przyk≈Çad:**  
Wektor **(3, 4)** w przestrzeni 2D wskazuje przemieszczenie o 3 jednostki w prawo oraz 4 jednostki w g√≥rƒô.

```python
# Let's define a 2D vector
v = np.array([3, 4])
print("Vector v:", v)

# Vector length (norm)
length = np.linalg.norm(v)
print("Length of v:", length)
```

```python
def draw_vectors(vectors, colors=None):
    plt.figure()
    ax = plt.gca()
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_aspect('equal')

    origin = [0], [0]

    if colors is None:
        colors = ['r'] * len(vectors)

    for i in range(len(vectors)):
        v = vectors[i]
        ax.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=colors[i])

    plt.grid()
    plt.show()

# Try visualizing one vector
draw_vectors([v])
```

```python
v1 = np.array([2, 1])
v2 = np.array([1, 3])
v_sum = v1 + v2

print("v1:", v1)
print("v2:", v2)
print("Sum:", v_sum)

# Visualize them
draw_vectors([v1, v2, v_sum], colors=['r', 'b', 'g'])
```

```python
v = np.array([2, 3])
scalar = 2
v_scaled = scalar * v

print("Original vector:", v)
print("After multiplying by", scalar, ":", v_scaled)

draw_vectors([v, v_scaled], colors=['purple', 'orange'])
```

```python
v1 = np.array([2, 3])
v2 = np.array([4, -1])

dot_product = np.dot(v1, v2)

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Dot product:", dot_product)

draw_vectors([v1, v2], colors=['green', 'blue'])
```

### üî∑ Orthogonality of Vectors / Ortogonalno≈õƒá Wektor√≥w

#### üá¨üáß English:

Two vectors are **orthogonal** (perpendicular) if the **angle between them is 90¬∞**.  
In vector algebra, this happens when their **dot product equals zero**:

$$
\vec{A} \cdot \vec{B} = 0
$$

This means the vectors do not "lean" toward each other in any direction ‚Äî they form a **right angle**.

**Example:**

$$
\vec{A} \cdot \vec{B} = 1 \cdot 0 + 0 \cdot 1 = 0
$$

---

#### üáµüá± Polski:

Dwa wektory sƒÖ **ortogonalne** (prostopad≈Çe), je≈õli **kƒÖt miƒôdzy nimi wynosi 90¬∞**.  
W algebrze wektor√≥w oznacza to, ≈ºe ich **iloczyn skalarny jest r√≥wny zeru**:

$$
\vec{A} \cdot \vec{B} = 0
$$

To znaczy, ≈ºe wektory sƒÖ **prostopad≈Çe** ‚Äì nie wp≈ÇywajƒÖ na siebie w ≈ºadnym kierunku.

**Przyk≈Çad:**  
$$
\vec{A} \cdot \vec{B} = 1 \cdot 0 + 0 \cdot 1 = 0
$$



```python
from numpy import dot
from numpy.linalg import norm
from math import acos, degrees

v1 = np.array([1, 0])
v2 = np.array([0, 1])

dot_product = dot(v1, v2)
angle = degrees(acos(dot_product / (norm(v1) * norm(v2))))

print("Dot product:", dot_product)
print("Angle between vectors:", angle, "degrees")

draw_vectors([v1, v2], colors=['red', 'blue'])
```

### üìå Wektor jednostkowy / Unit Vector


#### üá¨üáß English:

A **unit vector** is a vector that has a **magnitude of exactly 1**.  
It represents **direction and orientation**, but not size.

To find a unit vector in the direction of a given vector, divide it by its magnitude:

$$
\vec{u} = \frac{\vec{A}}{|\vec{A}|}
$$

---

#### üáµüá± Polski:

**Wektor jednostkowy** to taki wektor, kt√≥rego d≈Çugo≈õƒá (modu≈Ç) wynosi dok≈Çadnie **1**.  
Wskazuje on tylko **kierunek i zwrot**, bez informacji o wielko≈õci (skali).

Wektor jednostkowy wyznacza siƒô poprzez podzielenie danego wektora przez jego d≈Çugo≈õƒá:

$$
\vec{u} = \frac{\vec{A}}{|\vec{A}|}
$$

```python
v = np.array([3, 4])
unit_v = v / norm(v)

print("Original:", v)
print("Unit vector:", unit_v)
print("Length of unit vector:", norm(unit_v))

draw_vectors([v, unit_v], colors=['gray', 'green'])
```

# üìò Wprowadzenie do Notacji Bra-Ket / Introduction to Bra-Ket Notation


```python
# Define |0‚ü© and |1‚ü© states as vectors
ket_0 = np.array([1, 0])
ket_1 = np.array([0, 1])

print("|0‚ü©:", ket_0)
print("|1‚ü©:", ket_1)

draw_vectors([ket_0, ket_1], colors=['green', 'blue'])
```

```python
superposition = (ket_0 + ket_1) / np.sqrt(2)

print("Superposition state (|0‚ü© + |1‚ü©)/‚àö2:", superposition)
print("Length of the state:", norm(superposition))

draw_vectors([ket_0, ket_1, superposition], colors=['green', 'blue', 'purple'])
```
