## 🧭 Liczby zespolone na płaszczyźnie / Complex Numbers on the Plane

🇬🇧 They can be visualized as a point (or vector) on a 2D plane:  
- x-axis = real part  
- y-axis = imaginary part
---
\\
🇵🇱 Można je przedstawić jako punkt (lub wektor) na płaszczyźnie:  
- oś x = część rzeczywista  
- oś y = część urojona



```python
import matplotlib.pyplot as plt

def draw_complex_number(z):
  plt.plot([0, z.real], [0, z.imag], marker='o')
  plt.xlim(-5, 5)
  plt.ylim(-5, 5)
  plt.grid(True)
  plt.title("Complex Number on Argand Plane")
  plt.xlabel("Real")
  plt.ylabel("Imaginary")
  plt.axhline(0, color='black')
  plt.axvline(0, color='black')
  plt.show()

z2 = 2 + 3j
draw_complex_number(z2)
```

# 🧠 Liczby zespolone – definicja / Complex Numbers – definition

## 🔢 What is a Complex Number?

🇬🇧 A complex number has the form:  
$z = a + bi$, where:  
- a – real part  
- b - imaginary part  
- i – the imaginary unit, such that \( $i^2 = -1$\)

---

\\
## 🔢 Czym jest liczba zespolona?  

🇵🇱 Liczba zespolona ma postać:  
$z = a + bi$, gdzie:  
- a – część rzeczywista  
- b – część urojona  
- i – jednostka urojona, taka że \( $i^2 = -1$ \)


```python
# Representation of complex numer in Python
z = 3 + 4j
print("Real part:", z.real)
print("Imaginary part:", z.imag)
```

```python
# Complex arthimetic
z1 = 2 + 1j
z2 = 1 + 0.5j

print("Sum:", z1 + z2)
print("Product:", z1 * z2)

# Drawing numbers
draw_complex_number(z1)
draw_complex_number(z2)
draw_complex_number(z1 + z2)
draw_complex_number(z1 * z2)
```

## 📏 Moduł i sprzężenie zespolone / Modulus and Conjugate

🇬🇧  
- Modulus (length): $|z| = \sqrt{a^2 + b^2}$   
- Conjugate: $\overline{z} = a - bi$

---
\\

🇵🇱
- Moduł (długość): $|z| = \sqrt{a^2 + b^2}$   
- Sprzężenie zespolone: $\overline{z} = a - bi$


---


🧠 QC: Moduł \( |a + bi|^2 \) to prawdopodobieństwo stanu. Conjugate jest używany w iloczynie skalarnym \( ⟨ψ|φ⟩ \).  
🧠 QC: Modulus \( |a + bi|^2 \) gives the probability of a quantum state. Conjugate is used in inner products \( ⟨ψ|φ⟩ \).

## 🔄 Forma trygonometryczna i wzór Eulera (dla chętnych) / Polar Form and Euler’s Formula (Optional/Advanced)

🇬🇧  
- Euler’s formula: $e^{i\theta} = \cos(\theta) + i\sin(\theta)$
- A complex number can be written as: $z = r \cdot e^{i\theta}$

---
\\

🇵🇱  
- Wzór Eulera: $e^{i\theta} = \cos(\theta) + i\sin(\theta)$
- Liczbę zespoloną można zapisać jako:  $z = r \cdot e^{i\theta}$


```python
import cmath
import numpy as np

z = 1 + np.sqrt(3)*1j

# Polar form
r, theta = cmath.polar(z)
print("Moduł (r):", r)
print("Argument (theta w radianach):", theta)
print("Argument (stopnie):", np.degrees(theta))

z_back = cmath.rect(r, theta)
print("Zrekonstruowana liczba:", z_back)
```
