## ðŸ§­ Liczby zespolone na pÅ‚aszczyÅºnie / Complex Numbers on the Plane

ðŸ‡¬ðŸ‡§ They can be visualized as a point (or vector) on a 2D plane:  
- x-axis = real part  
- y-axis = imaginary part
---
\\
ðŸ‡µðŸ‡± MoÅ¼na je przedstawiÄ‡ jako punkt (lub wektor) na pÅ‚aszczyÅºnie:  
- oÅ› x = czÄ™Å›Ä‡ rzeczywista  
- oÅ› y = czÄ™Å›Ä‡ urojona



```python
import matplotlib.pyplot as plt

def draw_complex_number(z):
  plt.plot([0, z.real], [0, z.imag], marker='o')
  plt.xlim(-5, 5)
  plt.ylim(-5, 5)
  plt.grid(True)
  plt.title("Complex Number on Argand Plane")
  plt.xlabel("Real")
  plt.ylabel("Imaginary")
  plt.axhline(0, color='black')
  plt.axvline(0, color='black')
  plt.show()

z2 = 2 + 3j
draw_complex_number(z2)
```

## ðŸ”¢ What is a Complex Number?

ðŸ‡¬ðŸ‡§ A complex number has the form:  
$z = a + bi$, where:  
- a â€“ real part  
- b - imaginary part  
- i â€“ the imaginary unit, such that \( $i^2 = -1$\)

---

\\
## ðŸ”¢ Czym jest liczba zespolona?  

ðŸ‡µðŸ‡± Liczba zespolona ma postaÄ‡:  
$z = a + bi$, gdzie:  
- a â€“ czÄ™Å›Ä‡ rzeczywista  
- b â€“ czÄ™Å›Ä‡ urojona  
- i â€“ jednostka urojona, taka Å¼e \( $i^2 = -1$ \)


```python
# Representation of complex numer in Python
z = 3 + 4j
print("Real part:", z.real)
print("Imaginary part:", z.imag)
```

```python
# Complex arthimetic
z1 = 2 + 1j
z2 = 1 + 0.5j

print("Sum:", z1 + z2)
print("Product:", z1 * z2)

# Drawing numbers
draw_complex_number(z1)
draw_complex_number(z2)
draw_complex_number(z1 + z2)
draw_complex_number(z1 * z2)
```

## ðŸ”„ Forma trygonometryczna i wzÃ³r Eulera (dla chÄ™tnych) / Polar Form and Eulerâ€™s Formula (Optional/Advanced)

ðŸ‡¬ðŸ‡§  
- Eulerâ€™s formula: $e^{i\theta} = \cos(\theta) + i\sin(\theta)$
- A complex number can be written as: $z = r \cdot e^{i\theta}$

---
\\

ðŸ‡µðŸ‡±  
- WzÃ³r Eulera: $e^{i\theta} = \cos(\theta) + i\sin(\theta)$
- LiczbÄ™ zespolonÄ… moÅ¼na zapisaÄ‡ jako:  $z = r \cdot e^{i\theta}$


```python
import cmath
import numpy as np

z = 1 + np.sqrt(3)*1j

# Polar form
r, theta = cmath.polar(z)
print("ModuÅ‚ (r):", r)
print("Argument (theta w radianach):", theta)
print("Argument (stopnie):", np.degrees(theta))

z_back = cmath.rect(r, theta)
print("Zrekonstruowana liczba:", z_back)
```
